```python
import pandas as pd 

df = pd.read_csv("https://github.com/suhasmaddali/Twitter-Sentiment-Analysis/raw/main/train.csv")

df = df[['text', 'sentiment']]

df.columns= ['tweet', 'sentiment']

df.dropna(inplace=True)

```

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
import string
nltk.download('stopwords')
nltk.download('wordnet')
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    text = re.sub(f"[{string.punctuation}]", "", text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    text = " ".join([word for word in text.split() if word not in stop_words])
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    text = " ".join([lemmatizer.lemmatize(word) for word in text.split()])
    # Remove extra whitespace
    text = re.sub(r"\s+", " ", text).strip()
    return text
 
# Apply preprocessing to tweets
df['cleaned_tweets'] = df['tweet'].apply(preprocess_text)
 
tweets = df['cleaned_tweets']
 
# Tokenize tweets
tokenized_tweets = [tweet.split() for tweet in df['cleaned_tweets']]
print("Tokenized Tweets:\n", tokenized_tweets)
 
from gensim.models import Word2Vec
 
# Initialize and train Word2Vec model
word2vec_model = Word2Vec(sentences=tokenized_tweets, vector_size=1000, window=5, min_count=1, sg=0)
 
# Build vocabulary and train the model
word2vec_model.build_vocab(tokenized_tweets)
word2vec_model.train(tokenized_tweets, total_examples=word2vec_model.corpus_count, epochs=10)
 
import numpy as np
 
# Function to convert tweet to vector
def tweet_to_vector(tweet, model):
    words = tweet.split()
    vectors = [model.wv[word] for word in words if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)
 
# Convert all tweets to vectors
tweet_vectors = np.array([tweet_to_vector(tweet, word2vec_model) for tweet in df['cleaned_tweets']])
print("Tweet Vectors:\n", tweet_vectors)
```
