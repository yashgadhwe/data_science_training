### Label Encoding for the columns

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
feature = {}
for k in df.columns:
    if df[k].dtype == 'O':
        df[str(k)+'_Encode'] = lb.fit_transform(df[k])
        gt = {}
        ke = df[[k,str(k)+'_Encode']].drop_duplicates()[str(k)+'_Encode'].values
        val = df[[k,str(k)+'_Encode']].drop_duplicates()[k].values
        for i in range(len(ke)):
            gt[val[i]] = str(ke[i])
        feature[k] = gt

feature
```
