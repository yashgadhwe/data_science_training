```python
import pandas as pd

# Updated raw GitHub URL
url = "https://github.com/singhrau0/Big-Data-Preprocessing/raw/main/superstore-subset.xlsx"

# Read Excel file into Pandas DataFrame
df = pd.read_excel(url)

# Now you can work with the DataFrame (df) as needed
df
```

```python
pd.set_option('display.max_column', None)
df
```

```python
gt = df[['Product Sub-Category', 'Sales','Profit']].groupby('Product Sub-Category')

gt.agg({'Sales':['sum'],'Profit':['sum']}).sort_values(by = ('Sales','sum'), ascending= False)
```

```python
# Compare Region wise total sales and profit of each region of each product category
gt = df[['Region','Product Category','Sales','Profit']].groupby(['Region','Product Category'])

gt.agg(['sum']).sort_values(by = ['Region',('Profit','sum')], ascending=[True, True])
```
