```python
import pandas as pd
from sklearn.impute import KNNImputer

# Load the dataset
url = "https://raw.githubusercontent.com/singhrau0/Big-Data-Preprocessing/main/Leads.csv"
df = pd.read_csv(url)

df
```

```python
pd.set_option("display.max_columns", None)

df
```

```python


df.isnull().sum()
```

```python
# Display the first few rows of the dataframe
# print("Original DataFrame:")
# print(df.head())

# Check for missing values in the dataset
# print("\nMissing values in each column:")
# print(df.isnull().sum())

# Choose a column to apply KNN imputation
column_to_impute = 'TotalVisits'  # Replace with your column name

# Create a DataFrame with only the selected column
data_to_impute = df[[column_to_impute]]

# Initialize KNNImputer
knn_imputer = KNNImputer(n_neighbors=5)

# Apply KNN imputation
imputed_data = knn_imputer.fit_transform(data_to_impute)

# Replace the original column with the imputed data
df[column_to_impute] = imputed_data

# Display the DataFrame after imputation
print("\nDataFrame after KNN imputation:")
print(df.head())

# Check for missing values again to confirm imputation
# print("\nMissing values in each column after imputation:")
# print(df.isnull().sum())
```
